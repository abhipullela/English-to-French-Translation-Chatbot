```python
from transformers import MarianTokenizer, MarianMTModel
import time
#imports required for the project
```

```python
model_name = 'Helsinki-NLP/opus-mt-en-fr' #required model name for the translation (required a lot of research :))
tokenizer = MarianTokenizer.from_pretrained(model_name) #required tokenizer for the tokenization prior to translation
model = MarianMTModel.from_pretrained(model_name) #required model import
print('Model loaded successfully!') #check for no error
```

```python
def translate_text(text):
    tokens = tokenizer(text, return_tensors="pt", padding=True)
    translated = model.generate(**tokens)
    return tokenizer.decode(translated[0], skip_special_tokens=True)
#function designated for the translation of text
```

```python
def typing_effect(text):
    for char in text:
        print(char, end='', flush=True)
        time.sleep(0.03)
    print()
  #fancy chatbot style typing effect - uses the time import
```

```python
print('English-to-French Chatbot (type exit to stop)')
while True:
    user_input = input('You: ')
    if user_input.strip().lower() == 'exit':
        print('Chatbot: Au revoir!')
        break
    output = translate_text(user_input)
    print('Chatbot:', end=' ')
    typing_effect(output)
  #main menu
```
